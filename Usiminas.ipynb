```python
Usiminas = '/content/drive/MyDrive/Exemplos/USIM5.SA.csv'
import pandas as pd
import matplotlib.pyplot as plt
usim5 = pd.read_csv(Usiminas)
usim5.Date = pd.to_datetime(usim5.Date)
usim5.head()
```

```python
usim5.info()
```

```python
figura1 = plt.figure(figsize = (25, 8))
plt.plot(usim5.Date, usim5.Close)
```

```python
figura1 = plt.figure(figsize = (10, 10))
eixo1 = figura1.add_axes([0,0,3,1])
eixo1.plot(usim5.Date, usim5.Close)
```

```python
figura1 = plt.figure(figsize = (20, 10))
eixo1 = figura1.add_axes([0,0,2,1])
eixo1.plot(usim5.Date,  usim5.Close, color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize = 50, pad = 30)
eixo1.set_ylabel('Valor por ao em R$', fontsize = 50)
eixo1.set_xlabel('Data', fontsize = 30)
eixo1.legend(['Valor por ao em R$'], loc = 'upper right', fontsize = 20)
eixo1.set_facecolor('#008080')
eixo1.grid(False)
```

```python
from datetime import datetime
figura1 = plt.figure(figsize = (15, 8))
eixo1 = figura1.add_axes([0,0,2,1])
eixo1.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize = 20, pad = 20)
eixo1.set_ylabel('Valor por ao em R$', fontsize = 20)
eixo1.set_xlabel('Data', fontsize = 20)
eixo1.legend(['Valor por ao em R$'], loc = 'upper right', fontsize = 20)
eixo1.set_facecolor('#008080')
eixo1.grid(False)
eixo1.set_xlim(datetime(2014,1,1), datetime(2014,1,1))
```

```python
figura1 = plt.figure(figsize = (15, 8))
eixo1 = figura1.add_axes([0,0,2,1])
eixo2 = figura1.add_axes([1,0.5,0.8,0.3])



eixo1.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize = 20, pad = 20)
eixo1.set_ylabel('Valor por ao em R$', fontsize = 20)
eixo1.set_xlabel('Data', fontsize = 20)
eixo1.legend(['Valor por ao em R$'], loc = 'upper right', fontsize = 20)
eixo1.set_facecolor('#008080')
eixo1.grid(False)
eixo1.set_xlim(datetime(2014,1,1), datetime(2014,1,1))


eixo2.plot(usim5.Date, usim5.Close, color = 'springgreen')
eixo2.set_title('2014 - 2018')
eixo2.set_xlabel('Data', fontsize = 20)
eixo2.set_xlim(datetime(2014,1,1), datetime(2016,1,1))
eixo2.set_facecolor('#5F9EA0')
```

```python
figura1 = plt.figure(figsize = (15, 8))
eixo1 = figura1.add_axes([0,0,2,1])
eixo2 = figura1.add_axes([1,0.5,0.8,0.3])

esquerda = usim5.Date < datetime(2014,1,1)
direita = usim5.Date > datetime(2016,1,1)

eixo1.plot(usim5.Date, usim5.Close, color = 'red')
eixo1.plot(usim5[esquerda]['Date'], usim5[esquerda]['Close'], color = 'springgreen')
eixo1.plot(usim5[direita]['Date'], usim5[direita]['Close'], color = 'springgreen')
eixo1.set_title('Usiminas - USIM5', fontsize = 20, pad = 20)
eixo1.set_ylabel('Valor por ao em R$', fontsize = 20)
eixo1.set_xlabel('Data', fontsize = 20)
eixo1.legend(['Valor por ao em R$'], loc = 'upper right', fontsize = 20)
eixo1.set_facecolor('#008080')
eixo1.grid(False)


eixo2.plot(usim5.Date, usim5.Close, color = 'red')
eixo2.set_title('2014 - 2016', fontsize = 20, color = 'white')
eixo2.set_xlim(datetime(2014,1,1), datetime(2014,1,1))
eixo2.set_facecolor('#5F9EA0')
```

```python
fig2 = plt.figure(figsize = (5, 4))
eixo3 = fig2.add_axes([0,0,1,1])

eixo3.hist(usim5.Close, edgecolor = "#000000", color = '#60e9fd')
eixo3.set_title('Histograma do preo das aes USIM5')
eixo3.set_xlabel('Preo das aes em R$')
eixo3.set_ylabel('Frequencia em fechamentos')
eixo3.set_facecolor('#72adbb')
```

```python
vol_ano = usim5[['Date',
                 'Volume']].groupby(pd.Grouper(key = 'Date', freq = 'y')).sum('Volume')

fig2 = plt.figure(figsize = (5, 4))
eixo4 = fig2.add_axes([0,0,1,1])

eixo4.bar(vol_ano.index, vol_ano.Volume, width = 300, edgecolor = '#71b584',
          color = '#599c6c')
eixo4.set_title('Volume negociado por ano')
eixo4.set_ylabel('Preo das aes em TRI R$')
eixo4.set_xlabel('Ano')
eixo4.set_facecolor('#baffcb')
```
